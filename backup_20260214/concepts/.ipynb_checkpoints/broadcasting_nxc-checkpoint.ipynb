## Testing the validity of broadcasting-based DFT cross-correlation

The goal of this notebook is to verify, through examples, that performing 2D to 3D cross-correlation using DFT and broadcasting is valid.

```python
import numpy as np
from matplotlib import pyplot as plt

# We will ultimately be working with 2 simulated volumes of size 4 (slow) x 5 (depth) x 7 (fast)
# First create a source from which we can draw these samples with arbitrary offsets

n_slow = 9 # must be 4, 6, or 9
n_depth = 5
n_fast = 7
np.random.seed(1234)
src = np.random.rand(30,30,30)

ref_vol = np.zeros((n_slow,n_depth,n_fast))
tar_vol = np.zeros((n_slow,n_depth,n_fast))

ds = 0
dd = 0
df = 0

ref_vol[:] = src[:n_slow,:n_depth,:n_fast]
tar_vol[:] = src[ds:ds+n_slow,dd:dd+n_depth,df:df+n_fast]

plt.subplot(1,2,1)
plt.imshow(ref_vol[0,:,:])
plt.subplot(1,2,2)
plt.imshow(tar_vol[0,:,:])
```

```python
def show_layers(arr):
    nlayers = arr.shape[0]
    assert nlayers in [4,6,9]
    
    aarr = np.abs(arr)
    imax = np.max(aarr)
    imin = np.min(aarr)

    rows = int(np.ceil(np.sqrt(arr.shape[0])))
    cols = int(np.floor(np.sqrt(arr.shape[0])))
    
    for k in range(nlayers):
        plt.subplot(cols,rows,k+1)
        plt.imshow(aarr[k,:,:],clim=[imin,imax])
        plt.title(k)
    plt.show()

show_layers(tar_vol)
```

```python
# Test using a single B-scan from the target
meps = 1e-20

for s in range(n_slow):
    ref_vol[s,:,:] = (ref_vol[s,:,:]-np.mean(ref_vol[s,:,:]))/np.std(ref_vol[s,:,:])
fref = np.conj(np.fft.fftn(ref_vol))

tar = tar_vol[8,:,:]
tar = (tar-np.mean(tar))/np.std(tar)
ftar = np.fft.fftn(tar)
prod = ftar*fref
#prod = prod/(np.abs(prod)+meps)
nxc = np.abs(np.fft.ifftn(prod))
nxc = np.fft.fftshift(nxc,axes=[1,2])

show_layers(nxc)
```

```python
ds = 1
dd = 0
df = 0

tar_vol[:] = src[ds:ds+n_slow,dd:dd+n_depth,df:df+n_fast]

tar = tar_vol[0,:,:]
tar = (tar-np.mean(tar))/np.std(tar)
ftar = np.fft.fftn(tar)
prod = ftar*fref
#prod = prod/(np.abs(prod)+meps)
nxc = np.abs(np.fft.ifftn(prod))
nxc = np.fft.fftshift(nxc,axes=[1,2])

show_layers(nxc)
```
